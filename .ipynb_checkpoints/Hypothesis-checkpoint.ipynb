```python
import pandas as pd
import os
import statsmodels.api as sm
import numpy as np
```

```python
filepath = os.path.join('Resources','bikecounts.csv')
bikecounts_df = pd.read_csv(filepath)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head()
```

```python
bikecounts_df.info()
```

```python
def remove_char(value):
    for c in ['T', '(S)', ' ']:
        value = value.replace(c, '')
    return value.strip()
```

```python
bikecounts_df['Precipitation'] = bikecounts_df['Precipitation'].apply(remove_char)
bikecounts_df['Precipitation'] = bikecounts_df['Precipitation'].replace('',0).astype('float64')
bikecounts_df.head()
```

```python
bikecounts_df.info()
```

```python
bikecounts_df.head()
```

```python
X = bikecounts_df["Precipitation"]
y = bikecounts_df['Brooklyn Bridge']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
```

```python
print(results.summary())
```

> Hypothesis: The number of bikes crossing the Brooklyn Bridge is proportional to the Precipitation.

> Null Hypothesis: The number of bikes crossing the Brooklyn Bridge is independent of Precipitation.

From the results of the OLS regression we can see that the p value is 0.000463 which is < 0.05 and hence the Null hypothesis can be rejected.

We also see that the R-squared value is 0.359 which is a positive value which indicates that there exists a direct relation between the number of bikes crossing the Brooklyn Bridge and the Precipitation.
